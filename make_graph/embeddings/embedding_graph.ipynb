```python
from sklearn.manifold import TSNE 
import numpy as np
from matplotlib import pyplot as plt

import torch
import torchvision.transforms as transforms
import torch.nn as nn
```

```python
from sklearn.preprocessing import normalize
dataset = "Yelp2"
model = "DirectAU"
user_emb = np.load(f"./user_embedding/user_{model}_{dataset}.npy")
user_emb = normalize(user_emb)
user_label = np.load(f"./user_embedding/user_label_{model}_{dataset}.npy")
```

```python
user_emb[0]
```

```python
len(user_label)
```

```python
tsne = TSNE(n_components=2, random_state=0) 
cluster = np.array(tsne.fit_transform(np.array(user_emb)))
```

```python
pop_u = cluster[user_label==1]
unpop_u = cluster[user_label==0]
plt.scatter(pop_u[:, 0], pop_u[:, 1], color='red', s=0.05)
plt.scatter(unpop_u[:, 0], unpop_u[:, 1], color='blue', s=0.05)
```
